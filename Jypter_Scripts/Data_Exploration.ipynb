## Fist Pump Motion

Start by reading the first pump data and comparison plotting

```python
fist_pump = pd.read_csv("../Training_Data/fist_pump/fist_pump000.csv")
print ('Number of samples: '+ str(len(fist_pump)))
fist_pump_pairplot = sb.pairplot(fist_pump)
```

```python
fist_pump.columns
```

To seperate the gyro and accel I create subsets based on the appropriate columns. 

To use seaborn to plot all x,y,z data on the same plots, I melt the data with the deltatime in all cases 

Relplot the data as a line

```python
accel_fist_pump = fist_pump[['deltaTime', 'Acc_X', 'Acc_Y', 'Acc_Z']]
accel_melted_fist_pump = accel_fist_pump.melt('deltaTime',var_name='Accel''s', value_name='Meters/Sec^2')
gyro_fist_pump = fist_pump[['deltaTime','Gyro_X', 'Gyro_Y', 'Gyro_Z']]
gyro_melted_fist_pump = gyro_fist_pump.melt('deltaTime',var_name='Gyro''s',value_name='Rad/Sec')
```

```python
fist_pump_accel = sb.relplot(x="deltaTime", y="Meters/Sec^2", hue='Accel''s', data=accel_melted_fist_pump, kind='line')
fist_pump_accel.set(ylim=(-22,22))
fist_pump_gyro = sb.relplot(x="deltaTime", y="Rad/Sec", hue='Gyro''s', data=gyro_melted_fist_pump, kind='line')
fist_pump_gyro.set(ylim=(-5.5,5.5))
```

# PCA

We see the graphs are different, but I want to know what the most important features(Accel_x for example) are for each motion

```python
from sklearn import decomposition
from sklearn.decomposition import PCA 
```

```python
pca_fist = decomposition.PCA()
fist_pump_pca = pca_fist.fit_transform(fist_pump[['Acc_X', 'Acc_Y', 'Acc_Z', 'Gyro_X', 'Gyro_Y', 'Gyro_Z']])
print(pca_fist.explained_variance_ratio_)
fist_comps = pd.DataFrame(pca_fist.components_, columns=['Acc_X', 'Acc_Y', 'Acc_Z', 'Gyro_X', 'Gyro_Y', 'Gyro_Z'])
```

```python
sb.heatmap(fist_comps, annot=True)
```

```python
print(sum(pca_fist.explained_variance_ratio_[0:4])) #keep roughly 98 percent of the info from our data
```

It seems that around 71 percent of the useful data is in PCA component 0. PCA component 0 is stongly negatively correlated with our Accel Y axis, although the other accels are still useful. PCA Comp 1 (2 percent of our data) is mor stongly tied to Accel in the X and weakly in the Accel Y. PCA Component 2 (0.09 percent of our data) is strongly tied to Accel Z and weakly to Accel Y. 

What's the take away, for detecting Fist Pumps? Acceleration Y is slightly more useful than the other accelerations. The Accels and Gyro plots pretty much confirm this. 

Acceleration is more useful than Gyration, right? Maybe, but probably not! I haven't normalized the Accerlation values to the Gyroscope values. This skew says accelration is more important to determing motion than gyration. Which anyone who salsa dances knows, isn't true.
